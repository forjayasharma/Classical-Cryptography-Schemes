```python
import gmpy2
from gmpy2 import mpz
import random
gmpy2.get_context().precision=2050
rand_state = gmpy2.random_state(42)

def generate_prime(bit_count):
    temp = gmpy2.mpz_rrandomb(rand_state, bit_count)
    return gmpy2.next_prime(temp)
p=0
q=0
N=1
e=0
d=0
while(p==q):
    p=generate_prime(1024)
    q=generate_prime(1024)
e=generate_prime(1024)
print("private keys:",p,",",q,",",e)
N=gmpy2.mul(p-1,q-1)
d=gmpy2.invert(e,N)
print("public keys:",d,",",N)

def encrypt(mm,p,q):
    c=[]
    message=" "
    for x in mm:
        cc=gmpy2.powmod(mpz(x),mpz(e),mpz(gmpy2.mul(p,q)))
        c.append(cc)
        message+=(chr(gmpy2.powmod(cc,1,mpz(128))))
    print("ciphertext",message)
    return c

def decrypt(c,p,q):
    dec=[]
    for x in c:
        cc=gmpy2.powmod(mpz(x),mpz(d),mpz(gmpy2.mul(p,q)))
        dec.append(cc)
    dec1=""
    for y in dec:
        dec1+=(chr(y))
    print("decrypted message:",dec1)
    return dec
```
